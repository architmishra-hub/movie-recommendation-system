```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
credits.info()
```

```python
# merging the two dataframe into one single dataframe : movies = movies + credits
```

```python
movies = movies.merge(credits, on = "title")
```

```python
movies.info()
```

```python
movies.shape
```

```python
# selecting only relevant columns for further steps
# genres
# movie_id
# keywords
# overview
# cast
# crew
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
# we going to merge ['overview','genres','keywords','cast','crew'] these into one single column named tags
```

```python
movies.isnull().sum()
```

```python
# dropping null values
movies = movies.dropna()
```

```python
movies.shape
```

```python
movies.duplicated().sum()
```

```python
movies['genres'][0]
```

```python
# that is a list of dictionaries, we need list of strings
# {"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"} 
# ["Action","Adventure","Fantasy","Science Fiction"]
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# this is string of list consisting dictionaries
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
# same with keywords column
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# in cast , we are going to extract top three actors
def convert_cast(obj):
    L = []
    count = 1
    for i in ast.literal_eval(obj):
        if count <= 3:
            L.append(i["name"])
            count+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

```python
# from crew , we are only extracting the name of the director
def convert_crew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i["name"])
    return L
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)
```

```python
movies.head()
```

```python
#Key insight: We should remove spaces within each name or term in the genres, keywords, cast, 
#and crew columns so that multi-word names like “James Cameron” are treated as a single entity, 
#rather than being split into separate tokens such as “James” and “Cameron.”
```

```python
for col in ['genres','keywords','cast','crew']:
    movies[col] = movies[col].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# now converting "overview into a list of strings"
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies.head()
```

```python
# now joining all the five columns into one
movies["tags"] = movies["overview"]+movies["genres"]+movies["keywords"]+movies["cast"]+movies["crew"]
```

```python
movies["tags"][0]
```

```python
new_df = movies[["movie_id","title","tags"]].copy()
```

```python
new_df['tags'][0]
```

```python
# now converting each of tags into string for applying further techniques
new_df["tags"] = new_df["tags"].apply(lambda x : " ".join(i.lower() for i in x))
```

```python
new_df["tags"][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = "english")
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
#now [abilities', 'ability', 'able'] stemming these kind of words
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
# calcultaing cosine distance/similarity now
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie = movie.lower()
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True, key = lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
new_df['title'] = new_df['title'].apply(lambda x:x.capitalize())
```

```python
recommend("batman begins")
```
